### Population Issue

- **Claim 1**: Poor countries are having too many children, leading to an increase in the world population which will reach 9 billion in 20 years.
- **Claim 2**: Wealthy countries are having too few children, resulting in aging population problems.

- **Ridley's Message**: Both claims are incorrect. Birth rates in poor countries are steadily declining, while birth rates in wealthy countries are rebounding. Eventually, birth rates in all countries will converge around 2.1, stabilizing the global population.

## Engineer vs. Designer

- Conflict in Data Visualization: There has been a conflict between those who emphasize functionality and prefer a logical, scientific approach and those who emphasize emotional aspects and aesthetic elements, considering themselves artists.

- There exists a middle ground between the two groups, and the boundary distinguishing the two approaches is ambiguous.

- Generally, those who emphasize functionality and a logical approach often come from backgrounds in statistics, cartography, and computer science.
- On the other hand, those who emphasize emotional aspects and aesthetic elements typically have backgrounds in graphic design, fine arts, and journalism.

## Engineer vs. Designer

- Edward Tufte (1990): Yale University Professor Emeritus of Political Statistics, Tufte is considered one of the most influential theorists in the field of information design and visualization techniques.

- The highlight of Tufte's theory: rigorous minimalism
    - If a graph can convey a lot of information with minimal ink, then it is a good graph.
    - It should provide maximum inspiration in a short time by using the least amount of ink on a small area of paper.

- Data-ink ratio: the amount of ink used to present the data / the total amount of ink used to print the graphic

**Tufte's Comments**

- "It may well be the best statistical graphic ever drawn."
- "It presents six variables: the size of the army, its position on a two-dimensional plane, the direction of the army's movement, dates during the retreat from Moscow, and temperature."
- "You could create such a drawing once in a million times, but there are no principles for making such a beautiful graphic. It represents minimalism."

**Why is it an excellent graph?**

- Until recently, techniques for understanding data have relied heavily on scatter plots, bar graphs, and line plots.

## High-dimensional visualization

::: {.panel-tabset}

### Fig1 -- 규빈의 아이디어 (하수)

```python
import matplotlib.pyplot as plt
df_kor = pd.concat([pd.concat([df1,df2],axis=1).assign(gender='m'),df3.assign(gender='f')])
df_for = df4.rename({'g':'gender'},axis=1)
fig,ax = plt.subplots(2,2)
ax[0][0].plot(df_kor[df_kor.gender =="m"].w,df_kor[df_kor.gender =="m"].h,'o',alpha=0.5,markersize=3)
ax[0][0].set_title("Male/Korean")
ax[0][1].plot(df_kor[df_kor.gender =="f"].w,df_kor[df_kor.gender =="f"].h,'o',alpha=0.5,markersize=3)
ax[0][1].set_title("Female/Korean")
ax[1][0].plot(df_for[df_for.gender =="m"].w,df_for[df_for.gender =="m"].h,'o',alpha=0.5,markersize=3)
ax[1][0].set_title("Male/Foreign")
ax[1][1].plot(df_for[df_for.gender =="f"].w,df_for[df_for.gender =="f"].h,'o',alpha=0.5,markersize=3)
ax[1][1].set_title("Female/Foreign")
fig.tight_layout()
```

### Fig2 -- 미나드의 아이디어

```python
import seaborn as sns
df = pd.concat([df_kor.assign(nationality='korean'),df_for.assign(nationality='foreign')]).reset_index(drop=True)
sns.scatterplot(
    data=df,
    x='w',
    y='h',
    hue='gender',
    style='nationality',
    alpha=0.7,
    size='nationality',
    sizes=(100, 15)
)
```
